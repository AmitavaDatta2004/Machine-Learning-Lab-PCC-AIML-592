```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n=100
x=np.random.rand(n,1)
print(x)
y=3+4*x+0.5*np.random.randn(n,1)
print(y)

plt.scatter(x,y,marker="*",c="yellow",label="Data point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
```

```python
from sklearn.preprocessing import add_dummy_feature

x_b=add_dummy_feature(x,1)
print(x_b)
theta=np.linalg.inv(x_b.T@x_b)@x_b.T@y
print(theta)
```

```python
x_new=np.array([[1.,0.],[1.,1.]])
print(x_new)
```

```python
y_pred=x_new@theta
print(y_pred)
```

```python
plt.plot(x_new,y_pred,"r-")
# plt.scatter(x,y)
```
