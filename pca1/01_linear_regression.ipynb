```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n=100
x=np.random.rand(n,1)
print(x)
```

```python
y=3+4*x+0.5*np.random.randn(n,1)
print(y)
```

```python
plt.scatter(x,y,c="y",marker="*")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature
x_new=add_dummy_feature(x,1)
print(x_new)

theta=np.linalg.inv(x_new.T@x_new)@x_new.T@y
print(theta)

x_pred=np.array([[0],[1]])
x_pred_new=add_dummy_feature(x_pred,1)
print(x_pred_new)

y_pred=x_pred_new@theta
print(y_pred)
```

```python
plt.plot(x_pred,y_pred,"r-")
plt.scatter(x,y,c="yellow",alpha=0.5)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg=LinearRegression()
lin_reg.fit(x,y)
```
