```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
m=100
X=np.linspace(0,1,m)
print(X)
```

```python


t0=1
t1=2
y=t0+t1*X*X+0.2*np.random.randn(m)
print(y)
```

```python
plt.scatter(X,y)
plt.xlabel("Range of data point")
plt.show()
```

```python
lin_reg=LinearRegression()
# type(lin_reg)

X=X.reshape(-1,1)
# y=y.reshape(-1,1)
print(X)
print(y)
lin_reg.fit(X,y)
```

```python
print(lin_reg.coef_,lin_reg.intercept_)
```

```python
lin_reg.score(X,y)
```

```python
y_pred=lin_reg.predict(X)
print(y_pred)

plt.scatter(X,y,marker="*",c="y")
# plt.scatter(X,y_pred,marker="+",c="g")
plt.plot(X,y_pred,'r')
plt.show()
```

```python
error=y-y_pred
plt.scatter(y_pred,error)
```
